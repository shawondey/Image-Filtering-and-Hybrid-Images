```python
import numpy as np
import cv2
from scipy import ndimage
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def my_imfilter(image, kernel):
    
    # spatial dimension of image and kernel
    
    dimensioni = image.shape[:2]
    (i_h, i_w) = dimensioni

    dimensionk = kernel.shape[:2]
    (k_h,k_w) = dimensionk
    
    (pad_h, pad_w) = (k_h // 2, k_w // 2)
    
    output = np.zeros(image.shape)
    
    # Moving kernel across every pixel
    for y in range(pad_h, i_h - pad_h):
        for x in range(pad_w, i_w - pad_w):
            
            # If coloured, loop for colours.
            for colour in range(image.shape[2]):
                #center pixel
                center = image[
                    y - pad_h : y + pad_h + 1, x - pad_w : x + pad_w + 1, colour]
                
                output[y, x, colour] = (center * kernel).sum() / 255

    #return the result of the output
    return output


#Gaussian Kernel
#kernel_lowpass = np.array([[0.0625, 0.125, 0.0625], [0.125, 0.25, 0.125],[0.0625,0.125,0.0625]], dtype="float");
kernel_lowpass = np.array([[0.003906, 0.0156, 0.02343, 0.0156, 0.003906],
                        [0.0156, 0.0625, 0.09375, 0.0625, 0.0156],
                        [0.02343,0.09375,0.0625,0.140625,0.02343],
                        [0.0156,0.0625,0.09375,0.0625,0.0156],
                        [0.003906,0.0156,0.02343,0.0156,0.003906]], dtype="float");

#kernel_blur = np.ones((9, 9), dtype="float") * (1.0 / (9 * 9))

# kernel highpass filter
kernel_highpass = np.array((
    [-1, -1, -1],
    [-1, 8, -1],
    [-1, -1, -1]), dtype="int")
```

```python
image = mpimg.imread("einstein.bmp")
Con_Out_1 = my_imfilter(image, kernel_highpass)
cv2.imwrite("high_Pass_E.png",(Con_Out_1 + 0.5) * 255)
cv2.waitKey(0)
Highpass_E = cv2.imread('high_Pass_E.png')
plt.imshow(Highpass_E)
    
```

```python
image = mpimg.imread("marilyn.bmp")
Con_Out_2 = my_imfilter(image, kernel_lowpass)
cv2.imwrite("lowpass_m.png",Con_Out_2 * 255)
cv2.waitKey(0) 
Lowpass_M = cv2.imread('lowpass_m.png')
plt.imshow(Lowpass_M)
```

```python
hybrid_img= Con_Out_1 + Con_Out_2
cv2.imwrite("hybrid_img.png",hybrid_img*255)
cv2.waitKey(0)
f_image = cv2.imread('hybrid_img.png')

smooth = ndimage.gaussian_filter(f_image,1.5)
cv2.imwrite("Final_image.png",smooth)
cv2.waitKey(0)
Final_image = cv2.imread('Final_image.png')

plt.imshow(Final_image)


fig, axes = plt.subplots(1,5)
for ax in axes:
    ax.axis('off')

(ax0,ax1,ax2,ax3,ax4) = axes

ax0 = fig.add_axes((-0.75,0,1.2,1.2))
ax0.imshow(Final_image)
ax1 = fig.add_axes((0.07,0,0.6,0.6))
ax1.imshow(Final_image)
ax2 = fig.add_axes((0.48,0,0.3,0.3))
ax2.imshow(Final_image)
ax3 = fig.add_axes((0.69,0,0.15,0.15))
ax3.imshow(Final_image)
ax4 = fig.add_axes((0.8,0,0.08,0.08))
ax4.imshow(Final_image)
ax0.axis('off')
ax1.axis('off')
ax2.axis('off')
ax3.axis('off')
ax4.axis('off')

```
