## Q1:
<div style="text-align: justify">Explicitly describe image convolution: the input, the transformation, and the output.Why is it useful for computer vision </div>

## Answer:
<div style="text-align: justify">Image convolution is an image processing technique of modifying an image by convolving a small matrix called kernel with the original image to get the effects like blur, sharpen, emboss etc. This convolving technique involves sliding a kernel over original image from left to right and top to bottom one row and one column at a time, doing element wise multiplication and then summing the result. Here the kernel matrix (3*3,5*5 etc) is our input. When we apply this process in an image, sometimes we get a new image with reduced matrix than our original image (machine sees image as an array of pixel). In order to avoid this issue, we use a technique called padding where we increase the size of the input image in such a way that post convolution size of the image is same as the original size. For example, if our original image size is 5*5 and kernel size is 3*3, then we compute padding as follows

Padding = image size-kernel size+1.


That means we add 1 row to top and bottom and 1 column to left and right. Then the image size becomes 7*7. After that we apply convolution using 3*3 kernel which reduces the image size to 5*5 which is same as the original image size. 
 </div>

## Q2:
<div style="text-align: justify">What is the difference between convolution and correlation? Construct a scenario thatproduces a different output between both operations and show some images of the result.(you can use built in correlation functions and convolution functions in scipy here ifdesired). </div>

## Answer:
<div style="text-align: justify">Difference between convolution and correlation:
Convolution is a linear operation on an image, whereas correlation is a measure of similarity between two images.The basic difference between convolution and correlation is that the convolution process rotates the matrix by 180 degrees. Most of the time the choice of using the convolution and correlation is up to the preference of the users, and it is identical when the kernel is symmetrical.

Also, correlation is the measure of similarity of image with itself which has a different time lag between them.
    
 </div>

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from scipy import ndimage



# read image
image = cv2.imread('cat.bmp',0)
# prepare the filter
kernel = np.array([[1,1,1],[1,1,0],[1,0,0]])
# apply kernel to the original image using convolution filtering
dst_conv  = ndimage.convolve(image, kernel, mode='constant', cval=1.0)
# apply kernel to the original image using correlation filtering
dst_corr = cv2.filter2D(image, -1, kernel)
# concatenate images horizontally
result = np.concatenate((image, dst_corr, dst_conv), axis=1)
cv2.imwrite("result.png", result)
cv2.waitKey(0) 
convcor = cv2.imread('result.png')
plt.imshow(convcor)
```

## Q3:
<div style="text-align: justify">What is the difference between a high pass filter and a low pass filter in how they are constructed and what they do to the image? Please provide example kernels and outputimage </div>

## Answer:
<div style="text-align: justify">Difference between a highpass filter and Lowpass filter:
Convolution is a linear operation on an image, whereas correlation is a measure of similarity between two images.The basic difference between convolution and correlation is that the convolution process rotates the matrix by 180 degrees. Most of the time the choice of using the convolution and correlation is up to the preference of the users, and it is identical when the kernel is symmetrical.

Also, correlation is the measure of similarity of image with itself which has a different time lag between them. So basically, the low-spatial scale is obtained by filtering an image with a low-pass filter, and the high spatial scale is obtained by filtering an image with a high-pass filter.
    
 </div>

```python
import numpy as np
import cv2
from scipy import ndimage
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def my_imfilter(image, kernel):
    
    # spatial dimension of image and kernel
    
    dimensioni = image.shape[:2]
    (i_h, i_w) = dimensioni

    dimensionk = kernel.shape[:2]
    (k_h,k_w) = dimensionk
    
    (pad_h, pad_w) = (k_h // 2, k_w // 2)
    
    output = np.zeros(image.shape)
    
    # Moving kernel across every pixel
    for y in range(pad_h, i_h - pad_h):
        for x in range(pad_w, i_w - pad_w):
            
            # If coloured, loop for colours.
            for colour in range(image.shape[2]):
                #center pixel
                center = image[
                    y - pad_h : y + pad_h + 1, x - pad_w : x + pad_w + 1, colour]
                
                output[y, x, colour] = (center * kernel).sum() / 255

    #return the result of the output
    return output


#Gaussian Kernel
#kernel_lowpass = np.array([[0.0625, 0.125, 0.0625], [0.125, 0.25, 0.125],[0.0625,0.125,0.0625]], dtype="float");
kernel_lowpass = np.ones((7, 7), dtype="float") * (1.0 / (7 * 7));

#kernel_blur = np.ones((9, 9), dtype="float") * (1.0 / (9 * 9))

# kernel highpass filter
kernel_highpass = np.array((
    [-1, -1, -1],
    [-1, 8, -1],
    [-1, -1, -1]), dtype="int")


image = mpimg.imread("cat.bmp")
Con_Out_1 = my_imfilter(image, kernel_highpass)
cv2.imwrite("high_Pass_E.png",(Con_Out_1 + 0.5) * 255)
cv2.waitKey(0)
Highpass_E = cv2.imread('high_Pass_E.png')
plt.imshow(Highpass_E)
```

```python
image = mpimg.imread("dog.bmp")
Con_Out_2 = my_imfilter(image, kernel_lowpass)
cv2.imwrite("lowpass_m.png",Con_Out_2 * 255)
cv2.waitKey(0) 
Lowpass_M = cv2.imread('lowpass_m.png')
plt.imshow(Lowpass_M)
```
